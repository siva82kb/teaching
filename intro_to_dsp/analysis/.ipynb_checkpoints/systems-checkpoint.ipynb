```python
n = np.arange(-10, 11, 1)
x = (n + 3)
x[n >= 0] = 3
x = x * (np.abs(n) <= 3)

fig = figure(figsize=(15, 4))
ax = fig.add_subplot(111)
ax.stem(n, x)
ax.set_xlim(n[0], n[-1])
ax.set_ylim(-0.5, 3.5)
ax.set_xticks(n)
ax.tick_params(axis='both', which='major', labelsize=15)
ax.set_xlabel("$n$", fontsize=20)
ax.set_ylabel("$x[n]$", fontsize=20)
ax.grid(color='0.85', linestyle='-', linewidth=0.5)
plt.tight_layout()

fig.savefig("discsig.png", format="png", dpi=300)
fig.savefig("discsig.eps", format="eps", dpi=300)
```

```python
n = np.arange(-20, 21, 1)
x = (n >= 0) * (n <= 10)
N = 4
y = np.array([np.sum(x[i-N:i]) if i >= 3
              else np.sum(x[:i])
              for i, _ in enumerate(n)]) * (1 / N)

fig = figure(figsize=(15, 7))
ax = fig.add_subplot(211)
ax.axvline(0, lw=1, color="0.5", linestyle="dashed")
ax.stem(n, x)
ax.set_ylim(-0.5, 1.5)
ax.set_xticks(n)
ax.set_xlim(-5, 20)
ax.tick_params(axis='both', which='major', labelsize=15)
ax.set_xlabel("$n$", fontsize=20)
ax.set_ylabel("$x[n]$", fontsize=20)
ax.grid(color='0.85', linestyle='-', linewidth=0.5)

ax = fig.add_subplot(212)
ax.axvline(0, lw=1, color="0.5", linestyle="dashed")
ax.stem(n, y)
ax.set_ylim(-0.5, 1.5)
ax.set_xticks(n)
ax.set_xlim(-5, 20)
ax.tick_params(axis='both', which='major', labelsize=15)
ax.set_xlabel("$n$", fontsize=20)
ax.set_ylabel("$y[n]$", fontsize=20)
ax.set_title("$y[n] = (1/4) \cdot \sum_{k=0}^{3} x[n - k]$", fontsize=14)
ax.grid(color='0.85', linestyle='-', linewidth=0.5)

plt.tight_layout()

fig.savefig("memory.png", format="png", dpi=300)
fig.savefig("memory.eps", format="eps", dpi=300)
stem(n, y)
```
