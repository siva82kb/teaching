## Hysteresis.
---
### True Sensor Model

```python
def true_sensor_model_hysteresis(x, up=True, xmin=1, xmax=5, scale=1.0):
    if up:
        _val = (1 / (1 + np.exp(-scale * (x - 3.5))))
        _minv = (1 / (1 + np.exp(-scale * (xmin - 3.5))))
        _maxv = (1 / (1 + np.exp(-scale * (xmax - 3.5))))
    else:
        _val = (1 / (1 + np.exp(-scale * (x - 2.5))))
        _minv = (1 / (1 + np.exp(-scale * (xmin - 2.5))))
        _maxv = (1 / (1 + np.exp(-scale * (xmax - 2.5))))
    return (xmax - xmin) * (_val - _minv) / (_maxv - _minv) + xmin
```

```python
# True sensor model.
xmin, xmax = 1, 5
_x = np.arange(xmin, xmax, 0.01)
up = np.array([True] * len(_x) + [False] * len(_x))
x = np.hstack((_x, _x[::-1]))
y = np.array([true_sensor_model_hysteresis(_x, _up, scale=1.5)
              for _x, _up in zip(x, up)])
dx, dy = x[-1] - x[0], y[-1] - y[0]
```

```python
fig = figure(figsize=(6, 5))
ax = fig.add_subplot(1, 1, 1)
ax.plot(x, y, lw=2)
ax.set_xlabel(r"$\phi_t$", fontsize=24)
ax.set_ylabel(r"$\phi_a$", fontsize=24)
xticks(fontsize=18)
yticks(fontsize=18)
ax.grid(color='0.8', linestyle='--', linewidth=0.5)
ax.set_xlim(xmin - 0.2 * dx, xmax + 0.2 * dx)
ax.set_ylim(xmin - 0.2 * dx, xmax + 0.2 * dx)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(True)
ax.spines['bottom'].set_visible(True)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

plt.tight_layout()

fig.savefig(f"{outdir}/idealsensorio-hysteresis.png", format="png", dpi=300);
```

### Predicting measurand from sensor output with the straight line model

```python
# Fit sesnor model using the calibration data.
senmod_actual = polyfit(xt, ya, deg=1)
ypred1 = polyval(senmod_actual, x)
ypred2 = polyval(senmod_actual, xt)

# Get the estimated smoothness values.
xpred1 = (ya - senmod_actual[1]) / senmod_actual[0]
```
