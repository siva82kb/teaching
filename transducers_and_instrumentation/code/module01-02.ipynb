```python
dt = 0.01
_t = np.arange(0, 50, dt)
zeta = [0, 0.1, 1.0, 10.0]
wn = 1.0

fig = figure(figsize=(15, 6))
ax = fig.add_subplot(211)
# Undamped.
sys = ([1.0], [1.0, 0.0, 1.0])
t, h1 = signal.impulse(sys, X0=None, T=np.arange(0, 50, 0.01))
ax.plot(t, h1 / np.max(h1), lw=2, color="tab:blue", label="Undamped")
# Underdamped
sys = ([1.0], [1.0, 2 * zeta[1], 1.0])
t, h2 = signal.impulse(sys, X0=None, T=np.arange(0, 50, 0.01))
ax.plot(t, h2 / np.max(h2), lw=2, color="tab:red", label="Underdamped")
# Critically damped
sys = ([1.0], [1.0, 2 * zeta[2], 1.0])
t, h3 = signal.impulse(sys, X0=None, T=np.arange(0, 50, 0.01))
ax.plot(t, h3 / np.max(h3), lw=2, color="tab:green", label="Critically damped")
# Overdamped
sys = ([1.0], [1.0, 2 * zeta[3], 1.0])
t, h4 = signal.impulse(sys, X0=None, T=np.arange(0, 50, 0.01))
ax.plot(t, h4 / np.max(h4), lw=2, color="black", label="Overdamped")

# ax.plot(t, (t >= 0) * np.exp(-tau[1] * t), lw=2, color="tab:red", label=f"$\\tau = {tau[1]}sec$")
# ax.set_xlabel(r"$t (sec)$", fontsize=18)
ax.set_ylabel(r"$h(t)$", fontsize=18)
xticks(fontsize=18)
yticks(fontsize=18)
ax.set_xlim(0, 50)
ax.grid(color='0.8', linestyle='--', linewidth=0.5)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(True)
ax.spines['bottom'].set_visible(True)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.set_title("Impulse Response", fontsize=18)
ax.legend(frameon=False, fontsize=16, loc='center left', bbox_to_anchor=(1, 0.5))

ax = fig.add_subplot(212)
# Undamped.
ax.plot(t, dt * np.cumsum(h1), lw=2, color="tab:blue", label="Undamped")
# Underdamped
sys = ([1.0], [1.0, 2 * zeta[1], 1.0])
t, h2 = signal.impulse(sys, X0=None, T=np.arange(0, 50, 0.01))
ax.plot(t, dt * np.cumsum(h2), lw=2, color="tab:red", label="Underdamped")
# Critically damped
sys = ([1.0], [1.0, 2 * zeta[2], 1.0])
t, h3 = signal.impulse(sys, X0=None, T=np.arange(0, 50, 0.01))
ax.plot(t, dt * np.cumsum(h3), lw=2, color="tab:green", label="Critically damped")
# Overdamped
sys = ([1.0], [1.0, 2 * zeta[3], 1.0])
t, h4 = signal.impulse(sys, X0=None, T=np.arange(0, 50, 0.01))
ax.plot(t, dt * np.cumsum(h4), lw=2, color="black", label="Overdamped")

# ax.plot(t, (t >= 0) * np.exp(-tau[1] * t), lw=2, color="tab:red", label=f"$\\tau = {tau[1]}sec$")
ax.set_xlabel(r"$t (sec)$", fontsize=18)
ax.set_ylabel(r"$y(t)$", fontsize=18)
xticks(fontsize=18)
yticks(fontsize=18)
ax.set_xlim(0, 50)
ax.grid(color='0.8', linestyle='--', linewidth=0.5)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(True)
ax.spines['bottom'].set_visible(True)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.set_title("Step Response", fontsize=18)
ax.legend(frameon=False, fontsize=16, loc='center left', bbox_to_anchor=(1, 0.5))

plt.tight_layout()

fig.savefig(f"{outdir}/secondorder-response.png", format="png", dpi=300);
```
